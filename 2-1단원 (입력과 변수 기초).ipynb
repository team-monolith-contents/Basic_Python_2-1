## **변수란 무엇일까?**

* **변수 : 데이터를 저장하기 위한 공간**
> 데이터를 저장할 수 있는 메모리입니다. 저장한 값을 변경할 수 있습니다. 

* **변수의 종류**
> 숫자(정수는 int, 실수는 float), 문자열(str), 리스트(list), 딕셔너리 
> 정수, 실수, 문자열, 리스트, 딕셔너리 등을 변수의 "자료형"이라고 합니다.

* **변수의 종류 : 정수, 실수, 문자열**
> 정수(int), 실수(float), 문자열(str)은 가장 기초적인 자료형입니다.   
> 정수형 : 음수, 양수, 0. (수학적 정의로서의 '정수'와 동일합니다.)   
> 실수형 : 소수 형태로 표현되는 자료형입니다. 예를 들어, -1.3, -1.2, -0.1, 0.0, 0.1, 1.2, 1.3 같은 꼴입니다.   
> 문자열 : 큰 따옴표 혹은 작은 따옴표로 감싸진 형태입니다.

## **자료형과 사칙연산**

* 파이썬에서도 사칙연산을 할 수 있습니다. 아래의 사칙 연산 기호들을 참고하세요!
연산 | 설명 | 예시 | 결과
---|:---:|---|---
\+ | 더하기 | 7 + 2 | 9
\- | 빼기 | 7 - 2 | 5
\* | 곱하기 | 7 * 2 | 14
\/ | 나누기 | 7 / 2 | 3.5
\// | 몫 | 7 // 2 | 3
\% | 나머지 | 7 % 2 | 1

#### **예제 4 : 변수로 사칙연산하기**


```python
# 예제_4_변수로_사칙연산하기

A = 7 # 변수 A에 7을 저장합니다.
B = 2 # 변수 B에 2를 저장합니다.

print(A + B)  # A와 B를 더한 값을 구합니다.
print(A - B)  # A에서 B를 뺀 값을 구합니다.
print(A * B)  # A와 B를 곱한 값을 구합니다.
print(A / B)  # A에서 B를 나눈 값을 구합니다.
print(A // B) # A를 B로 나눈 몫을 구합니다.
print(A % B)  # A를 B로 나눈 나머지를 구합니다.
```

* 문자열도 더하고 곱할 수 있습니다. 
> **"+"연산자** : 문자열을 연결해주는 연산자입니다. (문자열+문자열)  
> **"\*"연산자** : 같은 문자를 여러 번 출력해주는 연산자입니다. (문자열*정수)

> **⚠️ 주의(WARNING):**
>
> **여러개의 문자열을 더해서 출력하는 것과 `print()` 함수에 변수들을 ,(반점:comma)로 분리하여 출력하는 것에는 차이가 있습니다.**
> 
> `print()` 함수에 변수들을 ,로 분리하여 출력하는 것은 `print()` 함수가 인자로 받은 변수들을 공백으로 구분하여 출력하는 것입니다.  
> 이것과는 다르게 여러개의 문자열을 더해서 출력하는 것은 문자열들을 `+` 연산으로 이어붙인 뒤 그 결과를 `print()` 함수를 이용해 출력하는 것입니다.  
> 문자열간의 `+` 연산은 출력에만 사용되는 것이 아니라 문자열을 이어 붙여야 하는 다양한 상황에 사용할 수 있습니다.

#### **예제 5 : 문자로 연산하기**


```python
# 예제_5_문자로_연산하기

a = '코들'   # 변수 a에 '코들'을 저장합니다.
b = '파이썬' # 변수 b에 '파이썬'을 저장합니다.

# 변수 a와 b를 + 로 더하여 출력하면 두 변수가 공백 없이 이어서 출력되고
# a와 b를 print 함수의 인자로 주어 출력하면 두 변수가 공백으로 구분되어 출력됩니다.
print(a + b) # a와 b를 이어붙인 문자열을 출력합니다.  
print(a, b) # a와 b를 공백으로 구분하여 출력합니다.
```
